## RÃ©gression linÃ©aire

Le nettoyage va ici consister Ã  travailler sur les 37 valeurs manquantes de la colonne "margin_low".
Pour cela, nous allons effectuer une rÃ©gression linÃ©aire multiple en utilisant la mÃ©thode des Moindres CarrÃ©s Ordinaires (OLS).

### Initialisation du modÃ¨le

Avant de commencer, nous devons crÃ©er deux nouveaux DataFrames : 

- "df_notnull" contenant toutes les lignes avec des valeurs non nulles pour la colonne "margin_low"
- "df_null" contenant toutes les lignes avec des valeurs nulles pour cette mÃªme colonne

```python
#Division des donnÃ©es en deux parties : une partie avec des valeurs non manquantes pour 'margin_low'et une partie avec des valeurs manquantes
df_notnull = df.dropna(subset=['margin_low'])
df_null = df[df['margin_low'].isnull()]
```

Ensuite, nous devons procÃ©der Ã  la crÃ©ation d'un ensemble d'entraÃ®nement (training set) et d'un ensemble de test (test set). La sÃ©paration en training set / test set nous permettra d'Ã©valuer les performances de notre modÃ¨le sur des donnÃ©es non vues (pour vÃ©rifier si le modÃ¨le gÃ©nÃ©ralise bien Ã  de nouvelles donnÃ©es).

Pour cela la mÃ©thode `train_test_split`  de la bibliothÃ¨que ` Scikit-learn` .

```python
#CrÃ©ation d'un training set et d'un test set pour les valeurs non manquantes
X = df_notnull.drop('margin_low', axis=1)
y = df_notnull['margin_low']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

Nous pouvons maintent crÃ©Ã©r un modÃ¨le de rÃ©gression linÃ©aire Ã  l'aide de la mÃ©thode des moindres carrÃ©s ordinaires (OLS) en utilisant la bibliothÃ¨que ` StatsModels` .

```python
#CrÃ©ation d'un objet OLS et ajustement du modÃ¨le sur le training set
model = ols("margin_low ~ is_genuine+diagonal+height_left+height_right+margin_up+length", data=X_train.join(y_train)).fit()
#Affichage des rÃ©sultats de la rÃ©gression
print(model.summary())
```

Le modÃ¨le montre une valeur Ã©levÃ©e du RÂ² (0,607), ce qui indique que les variables explicatives fournies expliquent environ 60% de la variation de la variable cible 'margin_low'. Cela peut Ãªtre considÃ©rÃ© comme un bon ajustement du modÃ¨le.

Cependant, les rÃ©sultats montrent que certaines variables explicatives ne sont pas significatives pour prÃ©dire la variable cible. 

En effet, les variables 'diagonal', 'height_left', 'height_right', 'length' ne sont pas significatives avec des valeurs de p> 0,05.

Nous devons donc ajuster notre modÃ¨le en supprimant les variables non significatives.

### Suppression des variables non significatives

Nous devons sÃ©lectionner les variables les plus importantes pour la prÃ©diction de la variable cible. 
Dans notre cas, les variables "is_genuine" et "margin_up" doivent Ãªtre utilisÃ©es pour la prÃ©diction de "margin_low".

Nous crÃ©ons donc un nouveau training set et test set avec seulement ces deux variables.

```python
#CrÃ©ation d'un training set et d'un test set pour les valeurs non manquantes
X = df_notnull[['is_genuine', 'margin_up']]
y = df_notnull['margin_low']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
#CrÃ©ation d'un objet OLS et ajustement du modÃ¨le sur le training set
model = ols("margin_low ~ is_genuine+margin_up", data=X_train.join(y_train)).fit()
#Affichage des rÃ©sultats de la rÃ©gression
print(model.summary())
```

Les rÃ©sultats montrent que le modÃ¨le linÃ©aire avec les variables 'is_genuine' et 'margin_up' explique 60,6% de la variation de la variable dÃ©pendante 'margin_low.' 

La variable 'is_genuine' a un effet nÃ©gatif significatif sur la variable dÃ©pendante, ce qui signifie que les billets authentiques ont une marge plus faible que les billets contrefaits. La variable 'margin_up' a Ã©galement un effet nÃ©gatif significatif.

### Analyse des donnÃ©es atypiques et influentes

AprÃ¨s avoir estimÃ© notre modÃ¨le de rÃ©gression linÃ©aire, nous devons maintenant l'analyser.

CommenÃ§ons par l'atypicitÃ© et l'influence Ã©ventuelle de certaines donnÃ©es. 
On pourra retirer les donnÃ©es atypiques et influentes afin d'amÃ©liorer notre modÃ¨le.

#### Calcul des leviers

Nous allons ici rÃ©aliser les tests Ã  un niveau $\alpha=5$% .

RÃ©cupÃ©rons $n$, le nombre d'individus de l'Ã©chantillon d'entrainement, et $p$, le nombre de variables.

```python
# DÃ©finition des paramÃ¨tres
alpha = 0.05
n = X_train.shape[0]
p = 3
```

```python
# Calcul des leviers et dÃ©finition du seuil
X_train["obs"] = np.arange(1, n+1)
X_train['levier'] = model.get_influence().hat_matrix_diag

seuil_levier = 2*p/n
```

```python
# ReprÃ©sentation graphique
plt.figure(figsize=(10,6))
plt.bar(X_train['obs'], X_train['levier'])
plt.xlabel('Observation')
plt.ylabel('Leviers')
plt.plot([0, 1025], [seuil_levier, seuil_levier], color='r')
plt.show()
```

```python
# Identifier les leviers supÃ©rieurs au seuil
leviers_influents = X_train[X_train['levier'] > seuil_levier].index
print(leviers_influents.shape[0], "leviers supÃ©rieurs au seuil.")
print(round(leviers_influents.shape[0] / X_train.shape[0] * 100,2),"% du training set.")
```

#### Calcul des rÃ©sidus studentisÃ©s

```python
# Calcul des rÃ©sidus studentisÃ©s et dÃ©finition du seuil
X_train['rstudent'] = model.get_influence().resid_studentized_internal
seuil_rstudent = t.ppf(1-alpha/2,n-p-1)
```

```python
# ReprÃ©sentation graphique
plt.figure(figsize=(10,6))
plt.bar(X_train['obs'], X_train['rstudent'])
plt.xlabel('Observation')
plt.ylabel('RÃ©sidus studentisÃ©s')
plt.plot([0, 1025], [seuil_rstudent, seuil_rstudent], color='r')
plt.plot([0, 1025], [-seuil_rstudent, -seuil_rstudent], color='r')
plt.show()
```

```python
# Identifier les rÃ©sidus studentisÃ©s supÃ©rieurs au seuil
rstudents_influents = X_train[np.abs(X_train['rstudent']) > seuil_rstudent].index
print(rstudents_influents.shape[0], "rÃ©sidus studentisÃ©s supÃ©rieurs au seuil.")
print(round(rstudents_influents.shape[0] / X_train.shape[0] * 100,2),"% du training set.")
```

#### Distance de Cook

```python
# Calcul des distances de Cook et dÃ©finition du seuil
influence = model.get_influence().summary_frame()

X_train['dcooks'] = influence['cooks_d']
seuil_dcook = 4/(n-p)
```

```python
# ReprÃ©sentation graphique
plt.figure(figsize=(10,6))
plt.bar(X_train['obs'], X_train['dcooks'])
plt.xlabel('Observation')
plt.ylabel('Leviers')
plt.plot([0, 1025], [seuil_dcook, seuil_dcook], color='r')
plt.show()
```

```python
# Identifier les observations influentes
dcooks_influents = X_train[X_train['dcooks'] > seuil_dcook].index
print(dcooks_influents.shape[0], "observations influentes.")
print(round(dcooks_influents.shape[0] / X_train.shape[0] * 100,2),"% du training set.")
```

#### Suppression des valeurs atypiques et influentes

AprÃ¨s avoir identifiÃ© les observations atypiques et influentes Ã  partir des leviers, des rÃ©sidus studentisÃ©s et des distances de Cook, la prochaine Ã©tape est de dÃ©terminer les index de ces observations pour pouvoir les supprimer de l'ensemble d'apprentissage.

```python
# Obtenir les index Ã  supprimer
index_a_supprimer = list(set(leviers_influents) & set(dcooks_influents)) + list(set(rstudents_influents) & set(dcooks_influents))
print(len(set(index_a_supprimer)),"observations atypiques et influentes supprimÃ©es.")

# Supprimer les valeurs de X_train
X_train_clean = X_train.drop(index_a_supprimer, axis=0)
y_train_clean = y_train.drop(index_a_supprimer, axis=0)

# Suppression des colonnes inutiles
X_train_clean = X_train_clean.drop(["obs", "levier", "rstudent", "dcooks"], axis=1)
```

### Ajustement du modÃ¨le

Nous pouvons maintenant effectuer une nouvelle rÃ©gression linÃ©aire multiple en utilisant les donnÃ©es nettoyÃ©es obtenues prÃ©cÃ©demment.

```python
#CrÃ©ation d'un objet OLS et ajustement du modÃ¨le sur le training set
model = ols("margin_low ~ is_genuine+margin_up", data=X_train_clean.join(y_train_clean)).fit()
#Affichage des rÃ©sultats de la rÃ©gression
print(model.summary())
```

La suppression des donnÃ©es atypiques et influentes a permis d'amÃ©liorer significativement notre modÃ¨le.

Toutefois, l'analyse des conditions de la rÃ©gression linÃ©aire (Ã  suivre) dÃ©montre Ã  ce stade que nos donnÃ©es ne sont pas homoscÃ©dastiques et que par consÃ©quent, la variance des rÃ©sidus n'est pas constante. 
 
On applique donc une transformation logarithmique Ã  la variable cible "y_train_clean" pour amÃ©liorer cela. 

Ensuite, on ajuste Ã  nouveau notre modÃ¨le sur le training set.

#### Transformation logarithmique de la variable cible

```python
#Application de la transformation logarithmique
y_train_clean_log = np.log(y_train_clean)

#CrÃ©ation d'un objet OLS et ajustement du modÃ¨le sur le training set
model = ols("margin_low ~ is_genuine+margin_up", data=X_train_clean.join(y_train_clean_log)).fit()
#Affichage des rÃ©sultats de la rÃ©gression
print(model.summary())
```

Le modÃ¨le linÃ©aire avec transformation logarithmique de la variable dÃ©pendante a un RÂ² de 0,668. Cela indique que ce modÃ¨le explique environ 66,8% de la variance de la variable dÃ©pendante. Il est lÃ©gÃ¨rement infÃ©rieur aprÃ¨s la transformation logarithmique de la variable dÃ©pendante mais il s'agit d'un transformation indispensable pour obtenir des rÃ©sidus homoscÃ©dastiques.

Tous les coefficients sont significativement diffÃ©rents de zÃ©ro avec une p-valeur infÃ©rieure Ã  0,001, ce qui suggÃ¨re que ces coefficients sont statistiquement significatifs.

Enfin, le test Durbin-Watson qui vÃ©rifie l'autocorrÃ©lation des rÃ©sidus indique qu'il n'y a pas d'autocorrÃ©lation significative.

### Validation du modÃ¨le

Afin de valider notre modÃ¨le, nous devons analyser les rÃ©sultats de notre rÃ©gression et vÃ©rifier que certaines conditions sont bien remplies:

- **NormalitÃ© des rÃ©sidus**: Les rÃ©sidus doivent suivre une distribution normale
- **HomoscÃ©dasticitÃ©**: Les rÃ©sidus doivent avoir une variance constante.
- **Pas de MulticollinÃ©aritÃ©**: Pas de forte corrÃ©lation entre les variables indÃ©pendantes
- **LinÃ©aritÃ©**: Il doit y avoir une relation linÃ©aire entre les variables dÃ©pendantes et indÃ©pendantes.

#### NormalitÃ© des rÃ©sidus

Nous allons procÃ©der au test des hypothÃ¨ses suivantes:

- **HypothÃ¨se nulle H0** : les donnÃ©es sont normalement distribuÃ©es
- **HypothÃ¨se alternative H1** : les donnÃ©es ne sont pas normalement distribuÃ©es

Nous allons pour cela utiliser le test de Shapiro-Wilk.
La statistique de test est calculÃ©e comme suit :

- La statistique de test renvoie un nombre compris entre 0 et 1, oÃ¹ une valeur proche de 1 indique que les donnÃ©es sont proches d'une distribution normale.

- La p-value indique la probabilitÃ© d'obtenir un rÃ©sultat aussi extrÃªme que celui observÃ© dans les donnÃ©es si H0 est vraie. Si la p-value est infÃ©rieure au niveau de test (fixÃ© Ã  0,05), on rejette H0.

CommenÃ§ons par effectuer une reprÃ©sentation graphique:

```python
# CrÃ©er la figure avec deux sous-graphiques
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))

# Tracer l'histogramme des rÃ©sidus sur le premier sous-graphique
sns.histplot(model.resid, kde=True, ax=axes[0])
axes[0].set_xlabel('Valeurs rÃ©siduelles', fontsize=12, fontweight='bold')
axes[0].set_ylabel('Nombre de rÃ©sidus', fontsize=12, fontweight='bold')
axes[0].set_title('Histogramme des RÃ©sidus', fontsize=16, fontweight='bold')

# Tracer le Q-Q plot des rÃ©sidus sur le deuxiÃ¨me sous-graphique
st.probplot(model.resid, plot=axes[1])
axes[1].set_title('Q-Q plot des rÃ©sidus', fontsize=16, fontweight='bold')

# Afficher la figure
plt.show()
```

```python
#Test de normalitÃ© des rÃ©sidus
stat, p = shapiro(model.resid)

print("RÃ©sultat du test de normalitÃ© des rÃ©sidus")
print('------------------------------------------')
print('Statistique de test :', stat)
print('p-value :', p)

if p<0.05:
    print("Rejet de l'hypothÃ¨se nulle H0")
else:
    print("L'hypothÃ¨se nulle H0 est acceptÃ©e")
```

<div class="alert alert-block alert-info">
    
Le test de normalitÃ© des rÃ©sidus indique que la distribution des rÃ©sidus n'est pas normale avec une p-value infÃ©rieure au niveau de significativitÃ© de 5%. 
    
Par consÃ©quent, l'hypothÃ¨se nulle selon laquelle les rÃ©sidus suivent une distribution normale est rejetÃ©e. 

NÃ©anmoins, l'observation des rÃ©sidus, le fait qu'ils ne soient pas trÃ¨s diffÃ©rents d'une distribution symÃ©trique, et le fait que l'Ã©chantillon soit de taille suffisante permettent de dire que **les rÃ©sultats obtenus par le modÃ¨le linÃ©aire gaussien ne sont pas absurdes**, mÃªme si le rÃ©sidu n'est pas considÃ©rÃ© comme Ã©tant gaussien.
    
</div>

#### HomoscÃ©dasticitÃ©

Ce test consiste Ã  vÃ©rifier la constance de la variation des rÃ©sidus. En cas d'hÃ©tÃ©roscÃ©dasticitÃ©, la variance de l'erreur n'est pas constante. Dans le cas contraire, (hypothÃ¨se d'homoscÃ©dasticitÃ©) la variance de l'erreur est constante.

Nous allons procÃ©der au test des hypothÃ¨ses suivantes:

- **HypothÃ¨se nulle H0** : HomoscÃ©dasticitÃ©
- **HypothÃ¨se alternative H1** : HÃ©tÃ©roscÃ©dasticitÃ©

Nous allons pour cela utiliser le test de Breusch Pagan.

```python
# Test de Breusch-Pagan
bp_test = het_breuschpagan(model.resid, model.model.exog)

# Affichage des rÃ©sultats
print("RÃ©sultats du test de Breusch-Pagan")
print('------------------------------------------')
print("Statistique de test : ", bp_test[0])
print("p-value : ", bp_test[1])

if bp_test[1]<0.05:
    print("Rejet de l'hypothÃ¨se nulle H0")
else:
    print("L'hypothÃ¨se nulle H0 est acceptÃ©e")
```

<div class="alert alert-block alert-info">
    
Le test montre une p-value de 0.74, ce qui indique que l'hypothÃ¨se nulle de l'homoscÃ©dasticitÃ© est acceptÃ©e. On peut donc conclure que la <b>variance des rÃ©sidus est constante<b/>.
    
</div>

Un VIF Ã©levÃ© indique que la variable est fortement corrÃ©lÃ©e avec les autres variables, ce qui peut rendre l'estimation instable et peu fiable.

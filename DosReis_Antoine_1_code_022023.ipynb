## RÃ©gression linÃ©aire

Le nettoyage va ici consister Ã  travailler sur les 37 valeurs manquantes de la colonne "margin_low".
Pour cela, nous allons effectuer une rÃ©gression linÃ©aire multiple en utilisant la mÃ©thode des Moindres CarrÃ©s Ordinaires (OLS).

### Initialisation du modÃ¨le

Avant de commencer, nous devons crÃ©er deux nouveaux DataFrames : 

- "df_notnull" contenant toutes les lignes avec des valeurs non nulles pour la colonne "margin_low"
- "df_null" contenant toutes les lignes avec des valeurs nulles pour cette mÃªme colonne

```python
#Division des donnÃ©es en deux parties : une partie avec des valeurs non manquantes pour 'margin_low'et une partie avec des valeurs manquantes
df_notnull = df.dropna(subset=['margin_low'])
df_null = df[df['margin_low'].isnull()]
```

Ensuite, nous devons procÃ©der Ã  la crÃ©ation d'un ensemble d'entraÃ®nement (training set) et d'un ensemble de test (test set). La sÃ©paration en training set / test set nous permettra d'Ã©valuer les performances de notre modÃ¨le sur des donnÃ©es non vues (pour vÃ©rifier si le modÃ¨le gÃ©nÃ©ralise bien Ã  de nouvelles donnÃ©es).

Pour cela la mÃ©thode `train_test_split`  de la bibliothÃ¨que ` Scikit-learn` .

```python
#CrÃ©ation d'un training set et d'un test set pour les valeurs non manquantes
X = df_notnull.drop('margin_low', axis=1)
y = df_notnull['margin_low']

X_train_reg_lin, X_test_reg_lin, y_train_reg_lin, y_test_reg_lin = train_test_split(X, y, test_size=0.3, random_state=42)

print("Taille du training set:", X_train_reg_lin.shape[0])
print("Taille du test set:", X_test_reg_lin.shape[0])
```

Nous pouvons maintent crÃ©Ã©r un modÃ¨le de rÃ©gression linÃ©aire Ã  l'aide de la mÃ©thode des moindres carrÃ©s ordinaires (OLS) en utilisant la bibliothÃ¨que ` StatsModels` .

```python
#CrÃ©ation d'un objet OLS et ajustement du modÃ¨le sur le training set
model_reg_lin = ols("margin_low ~ is_genuine+diagonal+height_left+height_right+margin_up+length", data=X_train_reg_lin.join(y_train_reg_lin)).fit()
#Affichage des rÃ©sultats de la rÃ©gression
print(model_reg_lin.summary())
```

Le modÃ¨le montre une valeur Ã©levÃ©e du RÂ² (0,607), ce qui indique que les variables explicatives fournies expliquent environ 60% de la variation de la variable cible 'margin_low'. Cela peut Ãªtre considÃ©rÃ© comme un bon ajustement du modÃ¨le.

Cependant, les rÃ©sultats montrent que certaines variables explicatives ne sont pas significatives pour prÃ©dire la variable cible. 

En effet, les variables 'diagonal', 'height_left', 'height_right', 'length' ne sont pas significatives avec des valeurs de p> 0,05.

Nous devons donc ajuster notre modÃ¨le en supprimant les variables non significatives.

### Suppression des variables non significatives

Nous devons sÃ©lectionner les variables les plus importantes pour la prÃ©diction de la variable cible. 
Dans notre cas, les variables "is_genuine" et "margin_up" doivent Ãªtre utilisÃ©es pour la prÃ©diction de "margin_low".

Nous crÃ©ons donc un nouveau training set et test set avec seulement ces deux variables.

```python
#CrÃ©ation d'un training set et d'un test set pour les valeurs non manquantes
X = df_notnull[['is_genuine', 'margin_up']]
y = df_notnull['margin_low']

X_train_reg_lin, X_test_reg_lin, y_train_reg_lin, y_test_reg_lin = train_test_split(X, y, test_size=0.3, random_state=42)

print("Taille du training set:", X_train_reg_lin.shape[0])
print("Taille du test set:", X_test_reg_lin.shape[0])
```

```python
#CrÃ©ation d'un objet OLS et ajustement du modÃ¨le sur le training set
model_reg_lin = ols("margin_low ~ is_genuine+margin_up", data=X_train_reg_lin.join(y_train_reg_lin)).fit()
#Affichage des rÃ©sultats de la rÃ©gression
print(model_reg_lin.summary())
```

Les rÃ©sultats montrent que le modÃ¨le linÃ©aire avec les variables 'is_genuine' et 'margin_up' explique 60,6% de la variation de la variable dÃ©pendante 'margin_low.' 

La variable 'is_genuine' a un effet nÃ©gatif significatif sur la variable dÃ©pendante, ce qui signifie que les billets authentiques ont une marge plus faible que les billets contrefaits. La variable 'margin_up' a Ã©galement un effet nÃ©gatif significatif.

### Analyse des donnÃ©es atypiques et influentes

AprÃ¨s avoir estimÃ© notre modÃ¨le de rÃ©gression linÃ©aire, nous devons maintenant l'analyser.

CommenÃ§ons par l'atypicitÃ© et l'influence Ã©ventuelle de certaines donnÃ©es. 
On pourra retirer les donnÃ©es atypiques et influentes afin d'amÃ©liorer notre modÃ¨le.

#### Calcul des leviers

Nous allons ici rÃ©aliser les tests Ã  un niveau $\alpha=5$% .

RÃ©cupÃ©rons $n$, le nombre d'individus de l'Ã©chantillon d'entrainement, et $p$, le nombre de variables.

```python
# DÃ©finition des paramÃ¨tres
alpha = 0.05
n = X_train_reg_lin.shape[0]
p = 3
```

#### HypothÃ¨se  2: NormalitÃ© des rÃ©sidus

Nous allons procÃ©der au test des hypothÃ¨ses suivantes:

- **HypothÃ¨se nulle H0** : les donnÃ©es sont normalement distribuÃ©es
- **HypothÃ¨se alternative H1** : les donnÃ©es ne sont pas normalement distribuÃ©es

Nous allons pour cela utiliser le test de Shapiro-Wilk.
La statistique de test est calculÃ©e comme suit :

- La statistique de test renvoie un nombre compris entre 0 et 1, oÃ¹ une valeur proche de 1 indique que les donnÃ©es sont proches d'une distribution normale.

- La p-value indique la probabilitÃ© d'obtenir un rÃ©sultat aussi extrÃªme que celui observÃ© dans les donnÃ©es si H0 est vraie. Si la p-value est infÃ©rieure au niveau de test (fixÃ© Ã  0,05), on rejette H0.

CommenÃ§ons par effectuer une reprÃ©sentation graphique:

```python
# CrÃ©er la figure avec deux sous-graphiques
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))

# Tracer l'histogramme des rÃ©sidus sur le premier sous-graphique
sns.histplot(model_reg_lin.resid, kde=True, ax=axes[0])
axes[0].set_xlabel('Valeurs rÃ©siduelles', fontsize=12, fontweight='bold')
axes[0].set_ylabel('Nombre de rÃ©sidus', fontsize=12, fontweight='bold')
axes[0].set_title('Histogramme des RÃ©sidus', fontsize=16, fontweight='bold')

# Tracer le Q-Q plot des rÃ©sidus sur le deuxiÃ¨me sous-graphique
st.probplot(model_reg_lin.resid, plot=axes[1])
axes[1].set_title('Q-Q plot des rÃ©sidus', fontsize=16, fontweight='bold')

# Afficher la figure
plt.show()
```

```python
#Test de normalitÃ© des rÃ©sidus
stat, p = shapiro(model_reg_lin.resid)

print("RÃ©sultat du test de normalitÃ© des rÃ©sidus")
print('------------------------------------------')
print('Statistique de test :', stat)
print('p-value :', p)

if p<0.05:
    print("Rejet de l'hypothÃ¨se nulle H0")
else:
    print("L'hypothÃ¨se nulle H0 est acceptÃ©e")
```

Par consÃ©quent, l'hypothÃ¨se nulle selon laquelle les rÃ©sidus suivent une distribution normale est rejetÃ©e. 

NÃ©anmoins, l'observation des rÃ©sidus, le fait qu'ils ne soient pas trÃ¨s diffÃ©rents d'une distribution symÃ©trique, et le fait que l'Ã©chantillon soit de taille suffisante permettent de dire que **les rÃ©sultats obtenus par le modÃ¨le linÃ©aire gaussien ne sont pas absurdes**, mÃªme si le rÃ©sidu n'est pas considÃ©rÃ© comme Ã©tant gaussien.

Le test de Durbin Watson calcule une statistique de test qui varie entre 0 et 4, oÃ¹ une valeur de 2 indique l'absence d'autocorrÃ©lation (les rÃ©sidus ne sont pas corrÃ©lÃ©s entre eux) et des valeurs plus proches de 0 ou 4 indiquent une forte autocorrÃ©lation (les rÃ©sidus sont trÃ¨s corrÃ©lÃ©s entre eux).

```python
def train_logistic_regression(X_train, y_train, X_test, alpha=0.05, drop_non_significant=True):
    # EntraÃ®ner le modÃ¨le initial de rÃ©gression logistique
    model = sm.Logit(y_train, X_train).fit()
    
    # Extraire les noms des variables explicatives
    variable_names = model.params.index[1:]
    
    # Calculer les p-values pour chaque variable explicative
    p_values = [model.wald_test(term) for term in variable_names]
    
    # CrÃ©er un tableau pour stocker les rÃ©sultats du test de Wald
    wald_test_results = pd.DataFrame(index=variable_names, columns=['coef', 'std err', 'z', 'P>|z|', 'significant'])
    
    # Parcourir les variables et stocker les rÃ©sultats du test de Wald dans le tableau
    for i, name in enumerate(variable_names):
        wald_test_results.loc[name, 'coef'] = model.params[i+1]
        wald_test_results.loc[name, 'std err'] = model.bse[i+1]
        wald_test_results.loc[name, 'z'] = model.tvalues[i+1]
        wald_test_results.loc[name, 'P>|z|'] = p_values[i].pvalue
        wald_test_results.loc[name, 'significant'] = (p_values[i].pvalue < alpha)
    print(wald_test_results)
    if drop_non_significant:
        # Obtenir les noms des variables non significatives
        non_significant_vars = wald_test_results[wald_test_results['significant'] == False].index
        
        # Supprimer les variables non significatives du DataFrame X_train
        X_train = X_train.drop(non_significant_vars, axis=1)
        X_test = X_test.drop(non_significant_vars, axis=1)
        
        # EntraÃ®ner Ã  nouveau le modÃ¨le de rÃ©gression logistique sur le nouveau DataFrame X_train
        model = sm.Logit(y_train, X_train).fit()
        
    # Retourner le modÃ¨le de rÃ©gression logistique entraÃ®nÃ© sur le DataFrame X_train modifiÃ©
    return model, X_train, X_test
```

```python
#On lance la fonction
model_reg_log, X_train, X_test = train_logistic_regression(X_train, 
                                                           y_train, 
                                                           X_test, 
                                                           alpha=0.05, 
                                                           drop_non_significant=True)
```

```python
# HypothÃ¨se nÂ°4: il n'y a pas de valeurs extrÃªmes aberrantes
# Nous allons utiliser le test de Cook pour identifier les valeurs aberrantes potentielles.

# Get influence measures
influence = logit_model.get_influence()

# Obtain summary df of influence measures
summ_df = influence.summary_frame()

# Filter summary df to Cook distance
diagnosis_df = summ_df.loc[:,['cooks_d']]

# Append absolute standardized residual values
diagnosis_df['std_resid'] = st.zscore(logit_model.resid_pearson)
diagnosis_df['std_resid'] = diagnosis_df.loc[:,'std_resid'].apply(lambda x: np.abs(x))

# Sort by Cook's Distance
diagnosis_df.sort_values("cooks_d", ascending=False)
diagnosis_df
```

```python
# HypothÃ¨se nÂ°5: Il existe une relation linÃ©aire entre les variables explicatives et le logit de la variable de rÃ©ponse.
# Nous allons utiliser le test de Box-Tidwell

X_train_lt = X_train.copy()  # lt = logit transform

# Define continuous variables
continuous_var = ['height_right', 'margin_low', 'margin_up', 'length']

# Add logit transform interaction terms (natural log) for continuous variables e.g. Age * Log(Age)
for var in continuous_var:
    X_train_lt[f'{var}:Log_{var}'] = X_train_lt[var].apply(lambda x: x * np.log(x)) #np.log = natural log

# Keep columns related to continuous variables
cols_to_keep = continuous_var + X_train_lt.columns.tolist()[-len(continuous_var):]

# Redefine independent variables to include interaction terms
X_lt = X_train_lt[cols_to_keep]
y_lt = y_train.copy()

# Add constant
X_lt_constant = sm.add_constant(X_lt, prepend=False)
  
# Build model and fit the data (using statsmodel's Logit)
logit_results = GLM(y_lt, X_lt_constant, family=families.Binomial()).fit()

# Display summary results
print(logit_results.summary())
```
